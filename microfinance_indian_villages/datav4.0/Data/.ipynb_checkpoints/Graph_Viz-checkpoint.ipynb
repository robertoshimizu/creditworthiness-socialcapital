## Plotting with Matplotlib

```python
import matplotlib.pyplot as plt
import networkx as nx
G = nx.dodecahedral_graph()
nx.draw(G)  # networkx draw()
plt.draw()  # pyplot draw()
```

```python
plt.figure(figsize=(18,12))
nx.draw_kamada_kawai(G)  # networkx draw()
plt.draw()  # pyplot draw()
```

```python
nx.draw_random(G)  # networkx draw()
plt.draw()  # pyplot draw()
```

```python
nx.draw_shell(G)  # networkx draw()
plt.draw()  # pyplot draw()
```

```python
nx.draw_spectral(G)  # networkx draw()
plt.draw()  # pyplot draw()
```

```python
nx.draw_spring(G)  # networkx draw()
plt.draw()  # pyplot draw()
```

```python
nx.draw_circular(G)  # networkx draw()
plt.draw()  # pyplot draw()
```

```python
plt.figure(figsize=(15,13))
plt.axis("off")
nx.draw_networkx(G)
```

```python
G = nx.dodecahedral_graph()
nodes = nx.draw_networkx_nodes(G, pos=nx.spring_layout(G), node_size=100, node_color='#1f78b4')
```

```python
G = nx.dodecahedral_graph()
edges = nx.draw_networkx_edges(G, pos=nx.spring_layout(G))
```
