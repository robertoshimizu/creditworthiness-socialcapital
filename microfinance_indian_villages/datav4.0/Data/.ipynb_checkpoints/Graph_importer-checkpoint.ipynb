### Functions to create Graphs 

```python
import time
def timeit(method):
    def timed(*args, **kw):
        ts = time.time()
        result = method(*args, **kw)
        te = time.time()
        if 'log_time' in kw:
            name = kw.get('log_name', method.__name__.upper())
            kw['log_time'][name] = int((te - ts) * 1000)
        else:
            print('%r  %2.2f ms' % \
                  (method.__name__, (te - ts) * 1000))
        return result
    return timed
```

```python
def edgie(file_adj):
    tema = file_adj.split('_')[1]
    key = int(file_adj.split('_')[3].split('.')[0])
    file_key = 'key_vilno_{}.csv'.format(key)
    file_key = os.path.join(path_key,file_key)
    key_vila = pd.read_csv(file_key,header=None)
    key_vila.columns=['pid']
    file_adj = os.path.join(path_adj, file_adj)
    df_vil = pd.read_csv(file_adj, header=None)
    m = df_vil.shape[0]
    for i in range(m):
            for j in range(m-i):
                if (df_vil[i][j+i] == 1):
                    pid1 = int(key_vila['pid'][i])
                    pid2 = int(key_vila['pid'][j+i])
                    G.add_edge(pid1, pid2, type=tema)
                    #print(pid1, pid2, tema)
    print("edgie",G.number_of_nodes(),G.number_of_edges(), hex(id(G)))
    return G #'{}_vilno_{}.csv - processed'.format(tema, key)
```

```python
# Loop for creating the graph, not repeating as the adjacent matrix is simetric
# However we assume direction line -> to columm, first comes first
@timeit
def create_graph(file):
    key = int(file.split('_')[2].split('.')[0])
    adjs = '_vilno_{}.'.format(key)      
    files_vil = [f for f in os.listdir(path=path_adj) if adjs in f and 'HH' not in f]
    global G
    G = nx.MultiDiGraph() #local variable
    print('Initializing vilno_{} at mem address:'.format(key),hex(id(G)))
    
    with concurrent.futures.ThreadPoolExecutor() as executor:
        adj_map = executor.map(edgie, files_vil)
    
    
    #pool = multiprocessing.Pool()
    #adj_map = pool.map(edgie, files_vil)
    #pool.close() 
    #pool.join()    
    
    #adj_map = map(edgie, files_vil)
    list(adj_map)
    print('Finalizing vilno_{} at mem address:'.format(key),hex(id(G)))
    print('Graph vilno_{} -  Num_Nodes: {} - Num_Edges: {}'.format(key, G.number_of_nodes(), G.number_of_edges()))
    os.chdir(path_save)
    nx.write_gml(G,'vilno{}.gml.gz'.format(key))
    del(G)
    sucesso = '_vilno_{}. processado'.format(key) 
    
    return sucesso
```

```python
key_vil = [f for f in sorted(os.listdir(path=path_key)) if 'HH' not in f]
key_vil = sorted(key_vil, key=lambda x: int(os.path.splitext(x.split('_')[2])[0]))
```

```python
import matplotlib.pyplot as plt
def vizu(G):
    spring_pos = nx.spring_layout(G)
    plt.figure(figsize=(20,14))
    plt.axis("off")
    nx.draw_networkx(G,pos = spring_pos, with_labels = False, node_size=35, node_color='#1f78b4')
    plt.show()
```
